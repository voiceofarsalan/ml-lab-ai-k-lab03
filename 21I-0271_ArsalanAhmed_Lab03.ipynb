### Working on the Iris Dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data  # Feature matrix
```

```python
X
```

```python
X_normalized = ((X - np.mean(X, axis=0)) / np.std(X,axis=0))
covarianceM=np.cov(X_normalized,rowvar=False) 
```

```python
covarianceM
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covarianceM)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
plt.figure(figsize=(10,10))
colorlist=['red','blue','green','yellow']
for i in range(len(eigenvalues)):
    plt.quiver(np.mean(X,axis=0)[0],np.mean(X,axis=0)[1],eigenvectors[0, i],eigenvectors[1, i], angles='xy', scale_units='xy', scale=2, color=colorlist[i], label=f'Eigenvector {i + 1}')
plt.xlim(1,9)
plt.ylim(1,7)
plt.xlabel('X-Axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Eigenvectors Representation',color="black")
plt.grid()
plt.show()
```
